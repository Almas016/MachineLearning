```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("student_scores.csv", delimiter=",")
```

```python
fig, ax = plt.subplots()
ax.grid()
ax.scatter(data['Hours'], data['Scores'])
ax.set_xlabel('Hours')#name column
ax.set_ylabel('Scores')#name row
```

```python
x = data['Hours'].values
y = data['Scores'].values

slope, intersept = 0, 0
MSE = 0 #MSE = (y1 - (mx1+b))**2 + y2 - (mx2+b))**2 +...+ yn - (m * xn + b))**2) / n
for i in range(len(x)):
    MSE += (y[i] - (slope * x[i] + intersept))**2
print(MSE / float(len(x)))
```

```python
numer, denom = 0, 0
xmean, ymean = np.mean(x), np.mean(y)
for i in range(len(x)):
    numer += (x[i] - xmean) * (y[i] - ymean)
    denom += (x[i] - xmean) ** 2
theta1 = numer / denom
theta0 = ymean - (theta1 * xmean)

print("Theta1(slope) = {0}, theta0(intercept) = {1}".format(theta1,theta0)) 
```

```python
max_x = np.max(x)
min_x = np.min(x)

# Calculating line values x and y
xline = np.linspace(min_x, max_x)
yline = theta0 + theta1 * xline

plt.plot(xline, yline)
plt.scatter(x, y)

plt.grid()
plt.show()
```
