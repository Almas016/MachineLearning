Answer:<br/>
Tanimoto score is used for vectors that can take on continuous values. <br/>
It is designed so that if the vector only takes values of 1 and 0. <br/>
Jaccards score is used for vectors that take discrete values, most often for binary values (1 or 0).<br/>
it works the same as Jaccards.<br/>
The most popular similarity measure for comparing chemical structures represented by means of fingerprints is the Tanimoto (or Jaccard) coefficient T. <br/>Two structures are usually considered similar if T > 0.85 (for Daylight fingerprints).<br/>
I would imagine you would use Tanimoto's when you have a 'mixed' vector <br/>that has some continuous valued parts and some binary valued parts.<br/>




```python
#Function to calculate Tanimoto similarity score
def tanimoto (list1, list2):
    onlyA = 0
    onlyB = 0
    bothAB = 0
    for i in range(len(list1)):
        if list1[i] == 1 and list2[i] == 0:
            onlyA+=1
        if list1[i] == 0 and list2[i] == 1:
            onlyB+=1
        if list1[i] == 1 and list2[i] == 1:
            bothAB+=1
    print('onlyA:',onlyA)
    print('onlyB:',onlyB)
    print('bothAB:',bothAB)
    return bothAB/(onlyA+onlyB+bothAB)
```

```python
#Test
A = [1,0,1,1,1,0,0,0]
B = [1,1,0,1,1,0,1,0]
result = tanimoto(A,B)
print('Result: ',result)
```
