# Reading and Understanding the Data

```python
data = pd.read_csv('newdata.csv')
data.head()
```

```python
data['test'] = data['Name'].str.findall('[A-Z][^A-Z]*')
```

```python
data
```

```python
df3 = data['test'].apply(pd.Series)
```

```python
df3.info()
```

```python
#data['Company'] = data['Company'].str.split().str[0]
data['Company'] = data['Company'].str.split().str[0]
```

```python
#data.drop(['Model'], axis=1)
data['Model'] = df3[df3.columns[1:]].apply(
    lambda x: ''.join(x.dropna().astype(str)),
    axis=1
)
```

```python
import matplotlib.pyplot as plt
def showpie(columnName):
    df_sym = pd.DataFrame(data[columnName].value_counts())
    df_sym.plot.pie(subplots=True,labels = df_sym.index.values, autopct='%1.1f%%', figsize = (15,10))
    # Unsquish the pie.
    plt.gca().set_aspect('equal')
    plt.show()
    plt.tight_layout()
    
showpie('Company')
```
