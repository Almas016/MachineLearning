# 1) Tanimoto score

```python
def tanimoto(a, b):
    a = ''.join(format(ord(i), 'b') for i in a) 
    a = list(a)
    
    b = ''.join(format(ord(i), 'b') for i in b) 
    b = list(b)

    aa = 0
    bb = 0
    c = 0
    
    if len(a) > len(b):
        for i in range(len(a) - len(b)):
            b.insert(0, 0)
        
    elif len(b) > len(a):
        for i in range(len(b) - len(a)):
            a.insert(0, 0)
        
    for i in range(len(b)):
        if int(a[i]) == int(b[i]) == 1:
            c += 1
        if int(a[i]) == 1 and int(b[i]) == 0:
            aa += 1
        if int(b[i]) == 1 and int(a[i]) == 0:
            bb += 1
    
    return str(c / (aa + bb + c)) + "%"
```

```python
a = "You should put me max mark"

b = "You sholdn't put me min mark"

print(tanimoto(b, a))
```

# 5) Audioscrobbler

```python
import requests
import pandas as pd
```

```python
API_KEY = '' #your api
USER_AGENT = '' #your username (Registered to)
```

```python
headers = {
    'user-agent': USER_AGENT
}

payload = {
    'api_key': API_KEY,
    'method': 'chart.gettopartists',
    'format': 'json'
}
```

```python
r = requests.get('http://ws.audioscrobbler.com/2.0/', headers=headers, params=payload)
r.status_code
```

```python
data = r.json()
data
```

```python
array = []
for i in data['artists']['artist']:
    array.append([i['name'], i['playcount'], i['mbid'], i['url'], i['streamable'], i['image']])
```

```python
df = pd.DataFrame(data=array, columns=["Name", "Playcount", "Mbid", "Url", "Streamable", "Image"])
```
